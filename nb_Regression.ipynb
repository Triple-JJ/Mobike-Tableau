```python
import numpy
import pandas as pd
from scipy import stats
import statsmodels.api as sm

df = pd.read_csv('orderid_regression_c.csv')
```

```python
df.head()
```

```python
df['intercept'] = 1
```

```python
df.head()
```

```python
lrm0 = sm.OLS(df['Total Records'], df[['intercept', 'Distinct count of bikeid']])
model0 = lrm0.fit()
model0.summary()
```

```python
lrm1 = sm.OLS(df['Total Records'], df[['intercept', 'Distinct count of userid']])
model1 = lrm1.fit()
model1.summary()
```

```python
lrm = sm.OLS(df['Total Records'], df[['intercept', 'max_bikeid', 'userid_growth']])
model = lrm.fit()
model.summary()
```

```python
lrm2 = sm.OLS(df['userid_growth'], df[['intercept', 'max_bikeid']])
model2 = lrm2.fit()
model2.summary()
```
